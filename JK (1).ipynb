```python
def find_outliers(ds,col):
    quart1=ds[col].quantile(0.25)
    quart3=ds[col].quantile(0.75)
    IQR=quart3-quart1
    low_val=quart1-1.5*IQR
    high_val=quart3+1.5*IQR   
    print(low_val)
    print(high_val)
    ds=ds.loc[(ds[col]<low_val)|(ds[col]<high_val)]
    return ds
data=pd.read_csv("auto-mpg.csv")
find_outliers(data,'acceleration')
                                  
```

```python
# !pip install wordcloud==1.8.1
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
alice_novel=open('alice_in_wonderland.txt')
alice_novel=alice_novel.read()
stopwords=set(STOPWORDS)
alice_wc=WordCloud(background_color='white',max_words=2000,stopwords=stopwords)

alice_wc.generate(alice_novel)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```
